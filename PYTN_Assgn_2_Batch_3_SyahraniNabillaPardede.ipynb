**Keterangan**

Selanjutnya adalah menghapus kolom EASE-MENT dan Unnamed: 0, dikarenakan data tersebut tidak diperlukan dan tidak ada nilai di dalam kolom tersebut, maka agar mendapatkan hasil yang akurat, maka 2 kolom tersebut dihapus dari tabel asli

```python
# Menghapus Kolom EASE-MENT dan Unnamed: 0, dikarenakan data tidak terlalu diperlukan
df_no_east = df.drop(["Unnamed: 0", "EASE-MENT"], axis=1)

# Cek Data
df_no_east.head(5)
```

**Keterangan**

Mengecek tipe data setiap kolom, agar lebih mempermudah untuk mengelompokkan setiap kolom dengan tipe data yang sama dan dengan tujuan untuk mempermudah dalam pengerjaan

```python
# Mengelompokan data dengan type data object
data_cat_obj = [
    "NEIGHBORHOOD", 
    "BUILDING CLASS CATEGORY", 
    "TAX CLASS AT PRESENT",
    "BLOCK",
    "LOT",
    "BUILDING CLASS AT PRESENT",
    "ADDRESS",
    "APARTMENT NUMBER",
    "BUILDING CLASS AT TIME OF SALE",
    "SALE DATE"
]
for data_group in data_cat_obj:
    df_no_east[data_group] = df_no_east[data_group].astype('category')

# Mengelompokan data dengan type data integer dan mengubahnya ke numeric
data_cat_int = [
    "BOROUGH",
    "SALE PRICE",
    "LAND SQUARE FEET",
    "GROSS SQUARE FEET",
    
]
for data_group in data_cat_int:
    df_no_east[data_group] = pd.to_numeric(df_no_east[data_group], errors='coerce')

# Mengubah type data SALE DATE dari Object menjadi Date
df_no_east["SALE DATE"] = pd.to_datetime(df_no_east["SALE DATE"])
```

**Catatan**

Untuk menghindari Error yang terjadi pada kolom SALE PRICE yang memiliki nilai (-), maka bisa mengabaikannya dengan menambahkan errors="coerce", yang dimana nantinya isi yang memiliki (-) akan diubah menjadi NaN

**More Information**

Parameters argscalar, list, tuple, 1-d array, or Series Argument to be converted.

errors{‘ignore’, ‘raise’, ‘coerce’}, default ‘raise’

* If ‘raise’, then invalid parsing will raise an exception.
* If ‘coerce’, then invalid parsing will be set as NaN.
* If ‘ignore’, then invalid parsing will return the input.

```python
mean_borough = 2.902477e+06
sd_borough = 2.141811e+07
n_borough = 18306
z = 1.96 # z-score untuk confidence level 95%
```

##Menghitung dengan memasukan rumus standard error

Rumus :

se = sd / np.sqrt(jumlah keseluruhan data)

**Keterangan:**

se = standard error
sd = standard devisiasi

```python
se = sd_borough / np.sqrt(n_borough)

# Cetak Hasil
print("Hasil standard error adalah:", se)
```

```python
lcb = mean_borough - z * se
ucb = mean_borough + z * se

print("Hasil Lower limit adalah:", lcb)
print("-------------------------")
print("Hasil Upper limit adalah:", ucb)
```

```python
# Grading Data
grade_borough_1 = df_no_east[df_no_east["BOROUGH"] == 1]
grade_borough_2 = df_no_east[df_no_east["BOROUGH"] == 3]

# Menghitung panjang nilai
# Data 1
num_grade_borough_1 = len(grade_borough_1)
# Data 2
num_grade_borough_2 = len(grade_borough_2)

# Menghitung Mean
# Data 1
mean_grade_borough_1 = grade_borough_1["SALE PRICE"].mean()
# Data 2
mean_grade_borough_2 = grade_borough_2["SALE PRICE"].mean()

# Menghitung Standard Deviasi
# Data 1
std_grade_borough_1 = grade_borough_1["SALE PRICE"].std()
# Data 2
std_grade_borough_2 = grade_borough_2["SALE PRICE"].std()
```

```python
print("Hasil perhitungan Data 1:",num_grade_borough_1, mean_grade_borough_1, std_grade_borough_1)
print("")
print("Hasil perhitungan Data 2:",num_grade_borough_2, mean_grade_borough_2, std_grade_borough_2)
```

```python
# Summary
sm.stats.ztest(grade_borough_1["SALE PRICE"].dropna(), grade_borough_2["SALE PRICE"].dropna(), alternative="two-sided")
```
